```python
n1, n2 = 0, 1

print("Fibonacci series till 52:")

while n1 <= 52:
    print(n1, end=" ")
    nth = n1 + n2
    n1 = n2
    n2 = nth
```

```python
def dfs(graph, start):
    visited = set()
    stack = [start] 
    while stack:
        vertex = stack.pop()        
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            for neighbor in reversed(graph[vertex]):
                if neighbor not in visited:
                    stack.append(neighbor)

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

dfs(graph,'A')
```
